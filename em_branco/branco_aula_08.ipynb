# Aula 8: Acessando vídeos de Youtube e PDFs

```python
import os
from langchain_groq import ChatGroq
from langchain.prompts import ChatPromptTemplate


api_key = 'SUA CHAVE DA GROQ AQUI'

os.environ['GROQ_API_KEY'] = api_key

chat = ChatGroq(model='llama-3.1-70b-versatile')
```

## Acessando o Youtube

E agora vamos mostrar o quão fácil é de acessar vídeos de youtube utilizando LangChain!

```python
from langchain_community.document_loaders import YoutubeLoader
url = 'https://www.youtube.com/watch?v=nLopLwffsLI&ab_channel=AsimovAcademy'

loader = YoutubeLoader.from_youtube_url(
    url,
    language=['pt']
)
lista_documentos = loader.load()
```

```python
documento = ''
for doc in lista_documentos:
    documento += doc.page_content
print(documento)
```

```python
template = ChatPromptTemplate.from_messages([
    ('system', 'Você é um assistente útil que possui as seguintes informações para formular uma resposta: {informacoes}'),
    ('user','{input}')
])
chain_yt = template | chat
resposta = chain_yt.invoke({'informacoes': documento, 'input': 'Segundo o autor, há quanto tempo ele ensina a programar?'})
print(resposta.content)
```

## Acessando PDFs

```python
from langchain_community.document_loaders import PyPDFLoader

caminho_pdf = '/home/fernando/fernando-ribeiro/Asimov_Academy/introducao_python_para_ia/arquivos/RoteiroViagemEgito.pdf'
loader = PyPDFLoader(caminho_pdf)
lista_documentos = loader.load()
documento = ''
for doc in lista_documentos:
    documento += doc.page_content

template = ChatPromptTemplate.from_messages([
    ('system', 'Você é um assistente útil que possui as seguintes informações para formular uma resposta: {informacoes}'),
    ('user','{input}')
])
chain_yt = template | chat
resposta = chain_yt.invoke({'informacoes': documento, 'input': 'Por quais cidades passaremos?'})
print(resposta.content)
```
