## Vari√°veis

**Uma vari√°vel √© uma forma de armazenar dados na mem√≥ria do computador para que possam ser utilizados e manipulados ao longo do c√≥digo.** Ela age como um "r√≥tulo" que aponta para um valor ou conjunto de dados.


### O que s√£o vari√°veis?

**Vari√°veis s√£o as caixas...**

Imagine que voc√™ tem **uma caixa** em sua casa, e essa caixa tem **um r√≥tulo**. **O r√≥tulo diz o que tem dentro da caixa**. Por exemplo, se voc√™ colocar "Doces" no r√≥tulo e dentro da caixa colocar balas, chocolates, etc., toda vez que voc√™ quiser doces, sabe que √© s√≥ olhar na caixa que tem esse r√≥tulo.

Na programa√ß√£o, uma vari√°vel √© como essa caixa com um r√≥tulo. Voc√™ d√° um **nome √† vari√°vel** (como o r√≥tulo da caixa) e coloca **um valor dentro dela** (como os doces na caixa). Ent√£o, sempre que voc√™ precisar desse valor, pode usar o nome da vari√°vel para encontr√°-lo.

Por exemplo:

- Nome da caixa: **nome**
- O que tem dentro da caixa: **"Adriano"**
Ent√£o, sempre que voc√™ quiser saber o nome, basta olhar para a vari√°vel nome, que est√° guardando o valor "Adriano".

Dessa forma, as vari√°veis ajudam a organizar e lembrar de informa√ß√µes no seu c√≥digo, como as caixas ajudam voc√™ a organizar seus objetos em casa.

**A mem√≥ria do computador √© a estante...**

A mem√≥ria do computador √© como a prateleira onde as caixas s√£o armazenadas. Quando voc√™ cria uma vari√°vel no seu programa, √© como se voc√™ pegasse uma caixa vazia na prateleira e colasse um r√≥tulo nela. Esse r√≥tulo permite que voc√™ saiba o que est√° guardado dentro da caixa sem precisar abri-la toda vez.

Por exemplo, ao criar a vari√°vel nome idade e armazena o valor "Adriano" nela, √© como se voc√™ colocasse uma caixa com o r√≥tulo nome na prateleira, e dentro dessa caixa, o valor "Adriano" est√° guardado. Sempre que o programa precisar saber o nome, ele vai at√© a prateleira, encontra a caixa com o r√≥tulo nome, e pega o valor que est√° dentro.

Na mem√≥ria do computador, essas caixas s√£o locais onde os dados s√£o armazenados temporariamente enquanto o programa est√° sendo executado. O r√≥tulo (nome da vari√°vel) facilita para o computador encontrar e usar os dados sem precisar lembrar de onde exatamente eles est√£o armazenados na prateleira (mem√≥ria).

Assim como na analogia da prateleira cheia de caixas, o computador organiza e recupera informa√ß√µes de forma eficiente usando essas "caixas" (vari√°veis) para manter tudo em ordem.

## Tipos de Dados

Tipos de dados s√£o as diferentes categorias que os valores podem ter em um programa. Cada tipo de dado define o que pode ser feito com aquele valor e como ele √© armazenado na mem√≥ria do computador.

### Principais tipos em Python
- *N√∫meros Inteiros (int)*: S√£o n√∫meros sem casas decimais, como 5, -2, ou 1000.

- *N√∫meros Decimais (float)*: S√£o n√∫meros que t√™m uma parte decimal, como 3.14, -0.5, ou 100.75. **Aten√ß√£o! Por padr√£o, Python utiliza a nota√ß√£o em ingl√™s, portanto, a casa decimal DEVE ser representada por ponto (.) e n√£o por v√≠rgula (,)!**

Por que √© importante?
Os tipos de dados s√£o importantes porque o computador precisa saber que tipo de valor ele est√° lidando para realizar opera√ß√µes corretamente. Por exemplo, voc√™ pode somar dois n√∫meros, mas n√£o pode somar uma palavra com um n√∫mero, pois s√£o tipos de dados diferentes.
