## Operadores de comparaÃ§Ã£o

Operadores de comparaÃ§Ã£o sÃ£o usados em programaÃ§Ã£o para comparar dois valores e determinar a relaÃ§Ã£o entre eles. O resultado da comparaÃ§Ã£o Ã© sempre um valor booleano, ou seja, True (verdadeiro) ou False (falso).

Aqui estÃ£o os principais operadores de comparaÃ§Ã£o:

- == (igual a): Verifica se dois valores sÃ£o iguais.
Exemplo: 5 == 5 resulta em True.

- != (diferente de): Verifica se dois valores sÃ£o diferentes.
Exemplo: 5 != 3 resulta em True.

- \> (maior que): Verifica se o valor Ã  esquerda Ã© maior que o valor Ã  direita.
Exemplo: 7 > 3 resulta em True.

- < (menor que): Verifica se o valor Ã  esquerda Ã© menor que o valor Ã  direita.
Exemplo: 2 < 5 resulta em True.

- \>= (maior ou igual a): Verifica se o valor Ã  esquerda Ã© maior ou igual ao valor Ã  direita.
Exemplo: 4 >= 4 resulta em True.

- <= (menor ou igual a): Verifica se o valor Ã  esquerda Ã© menor ou igual ao valor Ã  direita.
Exemplo: 3 <= 5 resulta em True.

Esses operadores sÃ£o fundamentais para tomar decisÃµes no cÃ³digo, como em condicionais (if) e loops, onde o programa precisa comparar valores para determinar qual aÃ§Ã£o executar a seguir.

```python
pergunta = input('Digite \'sair\' para sair: ')
print(pergunta.lower() == 'sair')
```

### Operadores lÃ³gicos

Operadores lÃ³gicos sÃ£o usados em programaÃ§Ã£o para combinar condiÃ§Ãµes ou inverter valores booleanos (True ou False). Eles ajudam a criar expressÃµes lÃ³gicas mais complexas. Aqui estÃ£o os principais operadores lÃ³gicos:

- and: Retorna True se todas as condiÃ§Ãµes forem verdadeiras.

> Exemplo: True and False resulta em False (porque nem todas as condiÃ§Ãµes sÃ£o verdadeiras).

- or: Retorna True se pelo menos uma das condiÃ§Ãµes for verdadeira.

> Exemplo: True or False resulta em True (porque pelo menos uma condiÃ§Ã£o Ã© verdadeira).

- not: Inverte o valor de uma condiÃ§Ã£o, transformando True em False e vice-versa.

> Exemplo: not True resulta em False.

Esses operadores sÃ£o usados para construir expressÃµes lÃ³gicas que controlam o fluxo do programa, permitindo que ele tome decisÃµes mais sofisticadas com base em mÃºltiplas condiÃ§Ãµes.

```python
pergunta = input('Digite \'sair\' para sair: ')
print(pergunta.lower() == 'sair' or pergunta.lower() == 'bye' or pergunta.lower() == 'exit')
```

## Estruturas condicionais if, elif e else

Estruturas condicionais sÃ£o usadas para tomar decisÃµes no cÃ³digo com base em condiÃ§Ãµes especÃ­ficas. Elas permitem que o programa execute diferentes blocos de cÃ³digo dependendo se uma condiÃ§Ã£o Ã© verdadeira ou falsa.

- if: Executa um bloco de cÃ³digo se a condiÃ§Ã£o for verdadeira.

- else: Executa um bloco de cÃ³digo se a condiÃ§Ã£o do if nÃ£o for verdadeira.

- elif: Verifica mÃºltiplas condiÃ§Ãµes, se a condiÃ§Ã£o do if nÃ£o for verdadeira.

```python
pergunta = input('Digite \'sair\' para sair: ')
if pergunta.lower() == 'sair' or pergunta.lower() == 'bye' or pergunta.lower() == 'exit':
  print('Saindo do programa...')
else:
  print('Continuar')
```
