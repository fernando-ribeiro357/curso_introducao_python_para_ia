# Aula 7: Dando accesso a sites para o nosso bot

O nosso objetivo é construir um bot que tenha acesso a informações personalizadas, que podem estar distribuídas em diversas fontes, como: sites, vídeos de Youtube, pdfs, etc.

A partir de agora, vamos dar opções ao nosso bot, para que ele consiga acessar as informações externas que ele utilizará como base para a geração de respostas mais específicas.

```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader("https://asimov.academy/")
lista_documentos = loader.load()
print(lista_documentos[0].page_content)
```
